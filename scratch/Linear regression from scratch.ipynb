## Linear Regression from scratch using Penguins dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pylab as plt
```

```python
df =pd.read_csv("Datasets/penguins.csv.txt")
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df.columns
```

```python
mat = df.corr()
ax= sns.heatmap(mat, annot = True)
_=plt.xticks(rotation =45)
```

```python
feature_name = "flipper_length_mm"
target_name = "body_mass_g"

scratch = pd.DataFrame()
scratch["Flipper Leangth (mm)"], scratch["Body Mass (g)"] = df[feature_name], df[target_name]
```

```python
_=sns.scatterplot(data = scratch, x ="Flipper Leangth (mm)",y= "Body Mass (g)", alpha = 0.5)
# representing the regressionn line
xline = np.linspace(scratch["Flipper Leangth (mm)"].min(),scratch["Flipper Leangth (mm)"].max(), num = 300)
yline = np.linspace(scratch["Body Mass (g)"].min(), scratch["Body Mass (g)"].max(), num = 300)
_=plt.plot(xline,yline,color = 'red')
```

```python
def linear_model_flipper_mass(flipper_length,weight_flipper_length,intercept_body_mass):
    """y=ax+b"""
    
    return weight_flipper_length*flipper_length + intercept_body_mass
```

```python
body_mass_180 = linear_model_flipper_mass(
    flipper_length=180, weight_flipper_length=40, intercept_body_mass=0)
body_mass_181 = linear_model_flipper_mass(
    flipper_length=181, weight_flipper_length=40, intercept_body_mass=0)

print(f"The body mass for a flipper length of 180 mm "
      f"is {body_mass_180} g and {body_mass_181} g "
      f"for a flipper length of 181 mm")
```
